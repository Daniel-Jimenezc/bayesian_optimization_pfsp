# Results boxplots

```python
df = pd.read_csv('taillard_results.csv')
df.head()
```

```python
df2 = pd.read_csv('taillard_results_sqexp.csv')
df2.head()
```

```python
row_name1 = 'instance'
row_name2 = 'kernel'
row_name3 = 'result'

df_tai20_5 = pd.DataFrame(columns=[row_name1, row_name2, row_name3])
df_tai20_10 = pd.DataFrame(columns=[row_name1, row_name2, row_name3])
df_tai20_20 = pd.DataFrame(columns=[row_name1, row_name2, row_name3])

df_tai20_5
```

```python
df2_tai20_5 = pd.DataFrame(columns=[row_name1, row_name2])
df2_tai20_10 = pd.DataFrame(columns=[row_name1, row_name2])
df2_tai20_20 = pd.DataFrame(columns=[row_name1, row_name2])
df2_tai20_5
```

```python
kernel = 'perm'
for index, row in df.iterrows():
    if 'tai20_5' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df_tai20_5 = df_tai20_5.append(aux)
    if 'tai20_10' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df_tai20_10 = df_tai20_10.append(aux)
    if 'tai20_20' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df_tai20_20 = df_tai20_20.append(aux)
```

```python
kernel = 'sqexp'
for index, row in df2.iterrows():
    if 'tai20_5' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df2_tai20_5 = df2_tai20_5.append(aux)
    if 'tai20_10' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df2_tai20_10 = df2_tai20_10.append(aux)
    if 'tai20_20' in row['file']:
        a = row['file'][19:-4]
        b = row['best_result']
        d = {'instance': a, 'kernel':kernel, 'result': b}
        aux = pd.DataFrame(data=d, index=[0])
        df2_tai20_20 = df2_tai20_20.append(aux)
```
